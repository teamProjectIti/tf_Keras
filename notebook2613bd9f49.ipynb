```python
import os
import cv2
import matplotlib.pyplot as plt
```

```python
image_folder = '/kaggle/input/catsvdogs-transformed/train_transformed'  # Update this if your folder name is different
image_files = os.listdir(image_folder)

# Optional: sort to ensure consistent order
image_files.sort()
```

```python
cat_images = [f for f in image_files if 'cat' in f.lower()]
dog_images = [f for f in image_files if 'dog' in f.lower()]
```

```python
def show_opencv_image(image_path, title=''):
    img = cv2.imread(image_path)               # Load image in BGR format
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # Convert to RGB for matplotlib
    plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Example: Show first 3 cat images
for img_name in dog_images[:3]:
    show_opencv_image(os.path.join(image_folder, img_name), title=img_name)
```

```python
data = []

# 0 = cat, 1 = dog
for f in cat_images:
    img = cv2.imread(os.path.join(image_folder, f))
    data.append((img, 0))

for f in dog_images:
    img = cv2.imread(os.path.join(image_folder, f))
    data.append((img, 1))
```
