```python
# Use a pipeline as a high-level helper
from transformers import pipeline

pipe = pipeline("text-classification", model="distilbert/distilbert-base-uncased-finetuned-sst-2-english")
```

```python
import torch
from transformers import DistilBertTokenizer, DistilBertForSequenceClassification

tokenizer = DistilBertTokenizer.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")
model = DistilBertForSequenceClassification.from_pretrained("distilbert-base-uncased-finetuned-sst-2-english")

inputs = tokenizer("love", return_tensors="pt")
with torch.no_grad():
    logits = model(**inputs).logits

predicted_class_id = logits.argmax().item()
model.config.id2label[predicted_class_id]
```

```python
classifier= pipeline("summarization", model="Falconsai/text_summarization")
text = (
    "I’m a Full Stack .NET Developer with 4 years of experience building modern web "
    "applications using ASP.NET Core and Angular. I’m passionate about creating clean, "
    "scalable solutions that make a real impact. My strengths include problem-solving, "
    "teamwork, and continuous learning. I value integrity, collaboration, and professional "
    "growth. Outside of work, I enjoy reading about technology and history. My goal is to "
    "keep improving my technical expertise and contribute to innovative projects that help others."
)
outputs=classifier(text)
# outputs
print(outputs[0]['summary_text'])

```

```python
synth = pipeline("text-to-audio", "facebook/musicgen-small")
text = "A relaxing track with Eastern influences, blended with touches of lofi, chillstep, and downtempo."

music = synth(text, forward_params={"do_sample":True})
```

```python
import scipy
scipy.io.wavfile.write("music.wav", rate=music["sampling_rate"], data=music['audio'])
```
