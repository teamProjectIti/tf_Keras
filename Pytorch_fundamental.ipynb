```python
import torch
import matplotlib.pyplot as plt

# Create random tensor image
random_tensor_image = torch.rand(size=(224, 224, 3))  # (H, W, C)

# Show shape and dimensions
print(random_tensor_image.shape, random_tensor_image.ndim)

# Display the image
plt.imshow(random_tensor_image)
plt.title("Random Tensor Image")
plt.axis("off")  # hides axes
plt.show()
```
