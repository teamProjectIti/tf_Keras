```python
from transformers import VitsModel, AutoTokenizer
import torch
import torchaudio

# تحميل النموذج والتوكينايزر
model = VitsModel.from_pretrained("facebook/mms-tts-ara")
tokenizer = AutoTokenizer.from_pretrained("facebook/mms-tts-ara")

# النص العربي اللي عايز تحوليه لصوت
text = "انا بيشوي امين من مصر "

# تحويل النص لإدخال
inputs = tokenizer(text, return_tensors="pt")

# توليد الصوت (مع تحديد الطول العشوائي للتنويع)
with torch.no_grad():
    output = model(**inputs).waveform

# حفظ الصوت كملف WAV
torchaudio.save("output_arabic.wav", output.squeeze(0).cpu(), sample_rate=16000)

print("الصوت اتحفظ في ملف output_arabic.wav")
```
